```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#a.Plot the Count of Famous Food Outlets / Restaurants in Bengaluru
zomato_df=pd.read_csv('Documents/zomato_restaurants.csv')
zomato_df
```

```python
#zomato_df['name'].value_counts()
plt.figure(figsize=(10, 7))
sns.set_style('darkgrid')
zomato_1= zomato_df['name'].value_counts()[:40]
sns.barplot(x= zomato_1, y = zomato_1.index, palette= 'BuGn_r')
sns.despine()
plt.title('Top 20 restaurant chains in Bangaluru')
plt.xlabel('Count')
plt.ylabel('Name of Restaurants')
```

```python
#b.Display accepting online orders vs not accepting online orders count % distribution
plt.figure(figsize = (8,8))
plt.pie(zomato_df['online_order'].value_counts(), labels = zomato_df['online_order'].unique(), autopct='%1.2f%%', explode = [0,0])
plt.legend()
plt.show()
```

```python
#c.Find the popular restaurant type in Bengaluru. 
plt.figure(figsize=(16,10))
a=sns.countplot(zomato_df['rest_type'])
plt.xticks(rotation=90)
```

```python
#d.Display accepting Table Booking vs Not accepting table booking count % distribution
plt.figure(figsize = (8,8))
plt.pie(zomato_df['book_table'].value_counts(), labels = zomato_df['book_table'].unique(), autopct='%1.2f%%', explode = [0,0])
plt.legend()
plt.show()
```

```python
#e.Histogram and Box Plot of Approx Cost Distribution. Note: Replace â€˜,â€™ from the values using replace(), lambda function with replace()
#e-Box Plot
zomato_1=zomato_df['approx_cost(for two people)'].str.replace(",", "")
#zomato_1
plt.figure(figsize=(11,8))
sns.boxplot(data = zomato_1, palette='Pastel1')
```

```python
#e-histogram

zomato_df['approx_cost(for two people)'].hist(bins = 15, figsize=(35, 10),color = 'darkorchid')
plt.title("Historam Plot of Approx_cost", fontsize = 12)
plt.xlabel("approx_cost(for two people)")
plt.ylabel("Frquency (in Range)")
plt.show()
```

```python
#f-histogram
zomato_df['rate'].hist(bins = 15, figsize=(35, 10),color = 'red')
plt.title("Rating distribution of the Restuarants", fontsize = 12)
plt.xlabel("Rate")
plt.ylabel("Frquency (in Range)")
plt.show()
```

```python
#g.Rate vs Approx Cost (for two people). Also categorise them into online_order available or not. 
sns.set_style('whitegrid')
plt.figure(figsize=(20,17))
sns.scatterplot(x = "rate", y = "approx_cost(for two people)", data = zomato_df, palette='viridis', hue = 'online_order')
plt.title('Cost vs Rating comparison')
plt.xlabel('Rating between 0 to 5')
plt.ylabel('Approx. cost for 2 people')
```

```python
#h.Distribution of Approx Cost for two people. Is it skewed?
zomato_1=zomato_df['approx_cost(for two people)'].str.replace(",", "")
plt.figure(figsize=(6,6))
sns.set_style('darkgrid')
sns.distplot(zomato_1, color= 'red')
plt.title('Approx. cost for two people in city of Banguluru')
plt.xlabel('Cost for two people')
```

```python
#i.Is there any difference b/w votes of restaurants accepting and not accepting online orders?
sns.set_style('whitegrid')
plt.figure(figsize=(9,6))
ax = sns.barplot(x = "online_order", y = "votes", data = zomato_df, ci = None, palette='Greens')
```

```python
#j.Which are the Top 20 most common restaurant types in Bengaluru?
#a=zomato_df['rest_type'].value_counts()
plt.figure(figsize=(10, 7))
sns.set_style('darkgrid')
zomato_1= zomato_df['rest_type'].value_counts()[:20]
sns.barplot(x= zomato_1, y = zomato_1.index, palette= 'flare')
sns.despine()
plt.title('Top 20 restaurant type in Bangaluru')
plt.xlabel('counts')
plt.ylabel('Type of Restaurants')
```

```python
#k.Which are the Top 20 most common Locations with most Number of Restaurants?
#a=zomato_df['location'].value_counts()
plt.figure(figsize=(10, 7))
sns.set_style('darkgrid')
zomato_1= zomato_df['location'].value_counts()[:20]
sns.barplot(x= zomato_1, y = zomato_1.index, palette= 'flare')
sns.despine()
plt.xlabel('count')
plt.ylabel('Location')
```
