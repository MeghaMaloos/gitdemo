```python
#4)Find the index positions of common in series A in another series B using numpy.where() and pd.Index() methods?
#using pd.Index
A = pd.Series([1, 3, 4, 5, 6, 7, 9, 10])
B = pd.Series([1, 3, 5, 7, 10])
result = [pd.Index(A).get_loc(i) for i in B]
result
```

```python
#using numpy.where()
A = pd.Series([9, 6, 4, 5, 3, 7,  1, 10,11])
B = pd.Series([1, 3, 5, 7, 10])
result = [np.where(i == A)[0].tolist()[0] for i in B]
result
```

```python
#6)How to filter valid emails from a series data using following pattern to be matched?
import re
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
emails = pd.Series(['books at amazom.com', 'james@egypt.com', 'jones@yahoo.in', 'bigdata@gmail.com'])
emails.str.findall(pattern,flags=re.IGNORECASE)
```
